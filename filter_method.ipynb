# Feature selection:
    
    Feature selection is a crucial process in machine learning that involves 
    identifying and selecting the most relevant input variables (features) for your model. 
    This technique helps you improve model performance, reduce overfitting, and enhance 
    generalization by eliminating redundant or irrelevant features.

## Importance of Feature Selection

Feature selection has a significant impact on the overall machine-learning process. Here are some key benefits:

### Improved Model Performance: 

    By focusing on the most relevant features, you can enhance the accuracy of your model in predicting new, unseen data.
### Reduced Overfitting: 

    Fewer redundant features mean less noise in your data, decreasing the chances of making decisions based on irrelevant information.
### Faster Training Times: 

    With a reduced feature set, your algorithms can train more quickly, which is particularly important for large-scale applications.
### Enhanced Interpretability: 

    By focusing on the most important features, you can gain better insights into the factors driving your models predictions.
### Dimensionality Reduction: 

    Feature selection helps to reduce the complexity of your model by decreasing the number of input variables.

| Method       | How it works                                                  | Pros                | Cons                                |
| ------------ | ------------------------------------------------------------- | -------------------- | ------------------------------------ |
| **Filter**   | Statistical tests (Chi-Square, ANOVA, Correlation)            | Fast, simple         | Ignores interaction between features |
| **Wrapper**  | Train model on different feature subsets                      | More accurate        | Slow, costly                         |
| **Embedded** | Model automatically selects important features (Lasso, Trees) | Efficient + powerful | Depends on model choice              |


## Information Gain

```python
from sklearn.feature_selection import mutual_info_classif
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
X, y = iris.data, iris.target

# Calculate information gain for each feature
info_gain = mutual_info_classif(X, y)

# Display results
for i, ig in enumerate(info_gain):
    print(f"Feature {i+1}: Information Gain = {ig:.4f}")
```

```python
info_gain
```

sklearn.feature_selection.mutual_info_classif

Function to calculate Mutual Information (Information Gain) between each feature and the target variable (for classification problems).

if the value is high we can keep it, if it is low we can remove it

## Chi2 method

```python
from sklearn.feature_selection import chi2
from sklearn.preprocessing import MinMaxScaler

# Normalize features to [0, 1] range for Chi-squared test
X_normalized = MinMaxScaler().fit_transform(X)

# Calculate Chi-squared scores
chi2_scores, _ = chi2(X_normalized, y)

# Display results
for i, score in enumerate(chi2_scores):
    print(f"Feature {i+1}: Chi-squared Score = {score:.4f}")
```

## missing value method

```python
import pandas as pd
import numpy as np

# Create a sample dataset with missing values
data = pd.DataFrame({
    'A': [1, 2, np.nan, 4, 5],
    'B': [np.nan, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, 5],
    'D': [1, 2, 3, 4, 5]
})

# Calculate missing value ratio
missing_ratio = data.isnull().mean()

# Display results
for feature, ratio in missing_ratio.items():
    print(f"Feature {feature}: Missing Value Ratio = {ratio:.2f}")
```

## Fisher's Score Method

```python
import numpy as np
from sklearn.datasets import load_iris

# Load Iris dataset
iris = load_iris()
X, y = iris.data, iris.target
```

```python
X
```

```python
# Fisher Score function
def fisher_score(X, y):
    n_features = X.shape[1]
    print(f"n_features: {n_features}")
    scores = []
    
    for i in range(n_features):  #range(4)  -- 0,1,2,3
        feature = X[:, i]    #X[:, 0]
        print(f"feature: {feature}")
        overall_mean = np.mean(feature)
        print(f"overall_mean: {overall_mean}")
        numerator = 0
        denominator = 0
        
        for cls in np.unique(y):
            print(f"cls: {cls}")
            cls_feature = feature[y == cls]    #feature[target_column == 0]
            print(f"cls_feature: {cls_feature}")
            cls_mean = np.mean(cls_feature)
            print(f"cls_mean: {cls_mean}")
            cls_var = np.var(cls_feature)
            print(f"cls_var: {cls_var}")
            n_cls = len(cls_feature)
            print(f"n_cls: {n_cls}")
            
            numerator += n_cls * (cls_mean - overall_mean) ** 2
            denominator += n_cls * cls_var
            print("numerator += n_cls * (cls_mean - overall_mean) ** 2")
            print(f"numerator += {n_cls} * ({cls_mean} - {overall_mean}) ** 2")
            print("************")
            print(f"numerator += {n_cls * (cls_mean - overall_mean) ** 2}")
            print("denominator += n_cls * cls_var")
            print(f"denominator += {n_cls} * {cls_var}") 
            print("************")  
            print(f"denominator += {n_cls* cls_var}")       
      
        scores.append(numerator / denominator if denominator != 0 else 0)
        print(f"Scores: {scores}")
    
    return np.array(scores)
```

```python
# Calculate Fisher Scores
fisher_scores = fisher_score(X, y)
```

```python
# Display results
for i, score in enumerate(fisher_scores):
    print(f"Feature {i+1}: Fisher Score = {score:.4f}")
```
