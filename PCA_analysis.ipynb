# Unsupervised Feature Selection Technique

## 1. Principal Component Analysis (PCA)

### PCA in Simple Terms

- Imagine you have data with many features (size, age, rooms, etc. in house price).
- Some features are highly correlated (e.g., ‚Äúsize‚Äù and ‚Äúnumber of rooms‚Äù).
- PCA finds new features (called principal components) that are:
- Combinations of the original features
- Capture the maximum variation in the data

Then, instead of keeping all features, we keep only the top few components (the ones with the most information).

üëâ In short:
- PCA = Compressing data into fewer dimensions, but keeping the most important patterns.

Real-Life Example

* You take a selfie ‚Üí the original image may have millions of pixels.
* You compress it ‚Üí file size reduces, but your face is still clear.
* PCA does the same: reduces features while keeping the essence of the data.

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
import numpy as np

# Load the Iris dataset
iris = load_iris()
X = iris.data

# Apply PCA (with all components)
pca = PCA()
pca.fit(X)

# Explained variance ratio
exp_var = pca.explained_variance_ratio_
print("Explained variance ratio:", exp_var)
cum_exp_var = np.cumsum(exp_var)
print("Cumulative explained variance:", cum_exp_var)

# Plot explained variance
plt.figure(figsize=(6,4))
plt.plot(range(1, len(exp_var)+1), cum_exp_var, marker='o', linestyle='--')
plt.title('Cumulative Explained Variance by PCA Components')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid(True)
plt.show()
```
